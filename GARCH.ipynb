```python
import pandas as pd
import numpy as np
import arch
from arch import arch_model
import os

import matplotlib.pyplot as plt

data_folder = 'data'
stock_files = [os.path.join(data_folder, file) for file in os.listdir(data_folder) if file.endswith('.csv')]
for stock in stock_files:
    stock_data = pd.read_csv(stock)
    stock_data['Date'] = pd.to_datetime(stock_data['Date'], format='%d/%m/%Y')
    stock_data.set_index('Date', inplace=True)
    stock_data['LogReturn'] = np.log(stock_data['Close'] / stock_data['Close'].shift(1))
    
    
    stock_data['LogReturn'] = stock_data['LogReturn'].replace([np.inf, -np.inf], np.nan)
    stock_data.dropna(subset=['LogReturn'], inplace=True)
    
    # Rescaling the 'LogReturn' to improve model parameter estimation
    stock_data['LogReturn'] = stock_data['LogReturn'] * 100
    
    garch_model = arch_model(stock_data['LogReturn'], vol='Garch', p=1, q=1)
    garch_result = garch_model.fit()
    print(garch_result.summary())
    
    time_periods = {
        '3M': '3M',
        '6M': '6M',
        '9M': '9M',
        'Full Period': None
    }
    
    for period, freq in time_periods.items():
        if freq:
    
            period_data = stock_data.loc[stock_data.index >= (stock_data.index.max() - pd.DateOffset(months=3*int(freq[0])))]
        else:
            period_data = stock_data
        
        period_log_return = period_data['LogReturn'].dropna()
        period_garch_model = arch_model(period_log_return, vol='Garch', p=1, q=1)
        period_garch_result = period_garch_model.fit(disp='off')
        
        conditional_volatility = period_garch_result.conditional_volatility
        
        mean_volatility = conditional_volatility.mean()
        
        annualized_volatility = mean_volatility * np.sqrt(252)  # Assuming 252 trading days in a year
        
        print(f"Stock: {stock[:-4]} - Volatility for {period}: {mean_volatility:.4f}")
        print(f"Stock: {stock[:-4]} - Annualized Volatility for {period}: {annualized_volatility:.4f}")


        plt.figure(figsize=(10, 6))
        plt.plot(stock_data.index, garch_result.conditional_volatility, label='Conditional Volatility')
        plt.title(f'GARCH Model Conditional Volatility for {os.path.basename(stock)[:-4]} - {period}')
        plt.xlabel('Date')
        plt.ylabel('Volatility')
        plt.legend()
        plt.show()

        # Rolling window forecast
        window_size = 252  # 1 year of trading days
        rolling_forecasts = []

        for start in range(len(stock_data) - window_size):
            end = start + window_size
            rolling_window_data = stock_data.iloc[start:end]
            rolling_log_return = rolling_window_data['LogReturn'].dropna()
            
            rolling_garch_model = arch_model(rolling_log_return, vol='Garch', p=1, q=1)
            rolling_garch_result = rolling_garch_model.fit(disp='off')
            
            forecast = rolling_garch_result.forecast(horizon=1)
            rolling_forecasts.append(forecast.variance.values[-1, 0] ** 0.5)

        if period == 'Full Period':
            # Align the forecasted volatility with the actual data
            rolling_forecasts = pd.Series(rolling_forecasts, index=stock_data.index[window_size:])

            # Plot actual vs forecast volatility
            plt.figure(figsize=(12, 6))
            plt.plot(stock_data.index, garch_result.conditional_volatility, label='Actual Volatility')
            plt.plot(rolling_forecasts.index, rolling_forecasts, label='Rolling Forecast Volatility', linestyle='--')
            plt.title(f'Actual vs Rolling Forecast Volatility for {os.path.basename(stock)[:-4]}')
            plt.xlabel('Date')
            plt.ylabel('Volatility')
            plt.legend()
            plt.show()

            # Forecast future volatility
            horizon = 14  # Forecasting Two Weeks Ahead
            forecasts = period_garch_result.forecast(horizon=horizon)
            forecast_volatility = forecasts.variance[-1:] ** 0.5

            # Plot the forecasted volatility
            plt.figure(figsize=(10, 6))
            plt.plot(forecast_volatility.T, marker='o', linestyle='-', label='Forecasted Volatility')
            plt.title(f'Forecasted Volatility for {os.path.basename(stock)[:-4]} - {period}')
            plt.xlabel('Days Ahead')
            plt.ylabel('Volatility')
            plt.legend()
            plt.show()
```
